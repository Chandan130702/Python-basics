```python
import keyword
word = input("Enter a word to check: ")
if keyword.iskeyword(word):
  print(f"'{word}' is a Python keyword.")
else:
  print(f"'{word}' is not a Python keyword.")
```

5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

```python
# Creating a tuple
my_list = [1, 2, 3, 4, 5]
print("Original List:", my_list)
my_list[2] = 10
print("Modified List:", my_list)

# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)
print("\nOriginal Tuple:", my_tuple)

# Attempting to modify tuple element (will raise an error)
try:
  my_tuple[2] = 10
except TypeError as e:
  print(f"Error: {e}")
  print("Tuples are immutable and cannot be modified.")
```

6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def modify_list(lst):
    lst.append(4)

def modify_number(n):
    n += 10

a = [1, 2, 3]
b = 5

modify_list(a)
modify_number(b)
# 4 will be added to the list that shows mutability
print("List after function:", a)
#number n will be same that shows immutability
print("Number after function:", b)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
sum = 0
for i in range(1, 51):
    if i % 2 == 0:
        sum += i
print("Sum of even numbers:", sum)
```

 14. Write a program to reverse a string using a while loop.

```python
text = input("Enter a string: ")
i = len(text) - 1

rev = ""
while i >= 0:
    rev += text[i]
    i -= 1

print("Reversed:", rev)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial =", fact)
```
